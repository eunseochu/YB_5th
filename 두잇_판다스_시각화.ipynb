# **04-2 matplotlib 라이브러리 자유자재로 사용하기**

#### **1) 기초 그래프 그리기** <br/>
seaborn 라이브러리의 tips 데이터 집합을 이용해 히스토그램, 산점도 그래프, 박스 그래프 그려보기 <br/>
- tips: 어떤 식당에서 팁을 지불한 손님의 정보를 모아둔 데이터 집합

> **1. 데이터 집합 불러오기** <br/>
tips 데이터 집합을 불러와 변수 tips에 저장
- 지불 금액, 팁, 성별, 흡연 유무, 요일, 시간, 전체 인원 정보를 담고 있음

데이터프레임의 열 데이터 분포와 빈도를 살펴보는 용도로 주로 사용

> **4. 히스토그램 그리기** <br/>
**hist 메서드**에 total_bill 열을 전달 <br/>
- x축의 간격은 bins 인잣값으로 조정 가능

> **5. 산점도 그래프란?** <br/>
변수 2개를 사용해서 만드는 그래프
- 변수 2개를 사용하기 때문에 **이변량 그래프**
- total_bill 열에 따른 tip 열의 분포를 나타낸 산점도 그래프

> **6. 산점도 그래프 그리기** <br/>
기본 틀과 그래프 격자를 만든 후, scatter 메서드에 total_bill, tips열 전달

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

> **7. 박스 그래프란?** <br/>
이산형 변수와 연속형 변수를 함께 사용하는 그래프 
- 이산형 변수: Female, Male처럼 명확하게 구분되는 값
- 연속형 변수: Tip과 같이 명확하게 셀 수 없는 범위의 값

> **8. 박스 그래프 그리기** <br/>
- 첫 번째 인자: 데이터를 추출하여 리스트에 담아 전달
- 두 번째 인자: labels 인잣값

```python
# tips 데이터프레임에서 성별이 Female인 데이터와 Maled인 데이터에서 tip 열 데이터만 추출하여 리스트에 담아 전달

boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'], 
               tips[tips['sex'] == 'Male']['tip']], 
               labels = ['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

#### **2) 다변량 그래프 그리기** <br/>
다변량 그래프는 3개 이상의 변수를 사용한 그래프 <br/>

> **1. 다변량 그래프 그리기 - 산점도 그래프** <br/>
앞서 실습한 산점도 그래프에 성별을 새로운 변수로 추가
- 성별은 색상으로 구분
- 그러나, 문자열은 그래프 색상을 지정하는 값으로 사용할 수 없으므로, 0이나 1 등의 정수를 사용
- 따라서 각 문자열은 **정수로 치환**하는 함수 필요

```python
# 여성일 경우 0을 반환하고 남성을 경우 1을 반환하는 함수

def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

> **2. sex_color 열 새로 생성** <br/>
recod_sex 함수가 반환한 값(0,1)을 데이터프레임에 sex_color 열로 추가
- recode_sex 함수를 **브로드캐스팅**하기 위해 apply 매서드 사용

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

> **3. size 변수 그래프에 추가** <br/>
테이블 당 인원 수(size)를 점의 크기로 표현
- scatter 메서드에 s, c 인자값으로 테이블당 인원 수와 성별의 치환값 전달
- s는 점의 크기(size)를, c는 점의 색상(color)을 의미

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size'] * 10,
    c = tips['sex_color'],
    alpha = 0.5) # 투명도 조절
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

#### **1) 단변량 그래프 그리기 - 히스토그램** <br/>
seaborn 라이브러리 이용 시 matplotlib보다 화려한 그래프를 그릴 수 있음

#### **2) 다양한 종류의 이변량 그래프 그리기** <br/>

> **1. seabron 라이브러리로 산점도 그래프 그리기** <br/>
matplotlib 라이브러리보다 다양한 방법으로 산점도 그래프 그릴 수 있음 
- seabron 라이브러리의 **regplot** 메서드 사용
- regplot 메서드 사용 시 산점도 그래프와 회귀선 함께 그리기 가능
- 회귀선 제거는 fit_reg인자를 **False**로 지정 시 가능

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips, fit_reg = False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

> **2. 산점도 그래프와 히스토그램 한 번에 그리기** <br/>
**jointplot** 메서드를 사용
- x, y인자에 원하는 열 이름을 지정하고 data 인잣값으로 데이터프레임 지정

```python
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips)

# 그래프 이름과 x, y축 이름 추가 가능
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

> **3. 육각 그래프(hexbin) 그리기** <br/>
2차원 표면에 육각형으로 데이터를 쌓아 표현하는 육각 그래프
- 산점도 그래프는 점이 겹쳐 보일 경우 점을 구분하기 어렵다는 단점이 있어, 육각 그래프를 사용하기도 함
- 특정 데이터의 개수가 많아지면 점점 진한 색으로 표현
- **jointplot** 메서드를 이용하며, kind 인잣값을 **hex**로 지정

```python
hexbin = sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "hex")
hexbin.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

> **4. 이차원 밀집도 그리기** <br/>
kdeplot 메서드로 만들 수 있음
- shade 인잣값을 **True**로 지정하면 그래프에 **음영 효과** 가능

```python
# total_bill, tip 열 데이터를 전달한 이차원 밀집도
ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'],
                 data2 = tips['tip'],
                 shade = True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

> **5. 바 그래프 그리기** <br/>

> **6. 박스 그래프 그리기** <br/>

> **7. 바이올린 그래프 그리기** <br/>

> **9. 관계 그래프 심화** <br/>
관계 그래프의 절반(대각선 기준)은 나머지 절반과 동일한 정보이므로 중복된 정보가 표현된다는 단점 존재
- 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체 가능
- **map_upper** 메서드로 대각선 기준 위쪽, **map_lower** 메서드로 아래쪽에 그릴 그래프 지정
- **map_diag**는 대각선을 중심으로 그래프 지정

```python
# 이차원 밀집도를 대각선 기준으로 아래에, 산점도 그래프는 위쪽에
# 히스토그램과 밀집도는 대각선을 중심으로

pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
plt.show()
```

# **04-5 seaborn 라이브러리로 그래프 스타일 설정**

> **1. 그래프에 스타일 적용하기** <br/>
>
> set_style 메서드를 사용하여 그래프 스타일 변경 가능
- 스타일을 적용할 부분에서 한 번만 실행하면, 메서드를 실행한 **이후의 그래프가 모두** 지정한 스타일로 변경됨
- darkgrid, whitegrid, dark, white, ticks 5종류

```python
fig, ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', hue = 'sex', data = tips, split = True)
```

> **2. whitegrid 스타일 설정** <br/>
그래프 배경에 가로 줄 생성

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', hue = 'sex', data = tips, split = True)
```

> **3. for문을 이용하여 모든 스타일 적용**

```python
fig = plt.figure()

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx + 1
  with sns.axes_style(style):
    ax = fig.add_subplot(2, 3, plot_position)
    violin = sns.violinplot(x = 'time', y = 'total_bill', data = tips, ax = ax)
    violin.set_title(style)

fig.tight_layout()
```
